```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set up plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Load the data directly from GitHub
github_url = 'https://raw.githubusercontent.com/Electricalelectronicsfinland/Introduction-to-Big-Data-Analytics-TX00EX95-3002/refs/heads/main/football_18-19-mATI.csv'

try:
    # Read the CSV file directly from GitHub
    df = pd.read_csv(github_url)
    
    print("‚úÖ Data loaded successfully from GitHub!")
    print(f"Dataset shape: {df.shape}")
    
except Exception as e:
    print(f"‚ùå Error loading file from GitHub: {e}")
    print("Please check:")
    print("1. The URL is correct and accessible")
    print("2. You have an active internet connection")
    exit()

# Display basic info about the dataset
print("\nüìä Dataset Overview:")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst few rows:")
print(df.head())
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())

# Convert Date to datetime if it's not already
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
else:
    print("‚ö†Ô∏è Warning: 'Date' column not found in dataset")

# Basic statistics for numerical columns
print("\nüìà Basic Statistics:")
numerical_cols = df.select_dtypes(include=[np.number]).columns
print(df[numerical_cols].describe())

# Check column names to ensure we have the expected football data columns
print("\nüîç Checking available columns:")
print(df.columns.tolist())

# Map common football data column names (adjust based on actual column names)
column_mapping = {
    'FTR': 'FTR',  # Full Time Result
    'HTR': 'HTR',  # Half Time Result
    'FTHG': 'FTHG', # Full Time Home Goals
    'FTAG': 'FTAG', # Full Time Away Goals
    'HTHG': 'HTHG', # Half Time Home Goals
    'HTAG': 'HTAG', # Half Time Away Goals
    'HS': 'HS',     # Home Shots
    'AS': 'AS',     # Away Shots
    'HST': 'HST',   # Home Shots on Target
    'AST': 'AST',   # Away Shots on Target
    'HF': 'HF',     # Home Fouls
    'AF': 'AF',     # Away Fouls
    'HC': 'HC',     # Home Corners
    'AC': 'AC',     # Away Corners
    'HY': 'HY',     # Home Yellow Cards
    'AY': 'AY',     # Away Yellow Cards
    'HR': 'HR',     # Home Red Cards
    'AR': 'AR',     # Away Red Cards
    'HomeTeam': 'HomeTeam',
    'AwayTeam': 'AwayTeam',
    'Referee': 'Referee'
}

# Check which columns we actually have
available_columns = {}
for standard_name, possible_name in column_mapping.items():
    if possible_name in df.columns:
        available_columns[standard_name] = possible_name
    else:
        print(f"‚ö†Ô∏è Column '{possible_name}' not found in dataset")

print(f"\n‚úÖ Available standard columns: {list(available_columns.keys())}")

# 1. Match Results Distribution (if we have the necessary columns)
print("\nüìä Creating Match Results Visualizations...")
plt.figure(figsize=(15, 12))

plot_count = 1

# Full Time Result distribution
if 'FTR' in available_columns:
    plt.subplot(3, 3, plot_count)
    result_counts = df[available_columns['FTR']].value_counts()
    colors = ['lightblue', 'lightcoral', 'lightgreen']
    labels = ['Home Win (H)', 'Away Win (A)', 'Draw (D)']
    
    # Ensure we have all three result types
    result_values = []
    result_labels = []
    for i, label in enumerate(labels):
        result_char = label.split(' ')[1].strip('()')
        if result_char in result_counts.index:
            result_values.append(result_counts[result_char])
            result_labels.append(label)
    
    if result_values:
        plt.pie(result_values, labels=result_labels, autopct='%1.1f%%', 
                colors=colors[:len(result_values)], startangle=90)
        plt.title('Full Time Result Distribution')
    plot_count += 1

# Half Time Result distribution
if 'HTR' in available_columns:
    plt.subplot(3, 3, plot_count)
    htr_counts = df[available_columns['HTR']].value_counts()
    htr_values = []
    htr_labels = []
    for i, label in enumerate(labels):
        result_char = label.split(' ')[1].strip('()')
        if result_char in htr_counts.index:
            htr_values.append(htr_counts[result_char])
            htr_labels.append(label)
    
    if htr_values:
        plt.pie(htr_values, labels=htr_labels, autopct='%1.1f%%', 
                colors=colors[:len(htr_values)], startangle=90)
        plt.title('Half Time Result Distribution')
    plot_count += 1

# Goals distribution
if all(col in available_columns for col in ['FTHG', 'FTAG']):
    plt.subplot(3, 3, plot_count)
    home_goals = df[available_columns['FTHG']]
    away_goals = df[available_columns['FTAG']]
    max_goals = max(home_goals.max(), away_goals.max())
    plt.hist([home_goals, away_goals], bins=range(0, max_goals + 2), 
             label=['Home Goals', 'Away Goals'], alpha=0.7)
    plt.xlabel('Goals')
    plt.ylabel('Frequency')
    plt.title('Goals Distribution')
    plt.legend()
    plot_count += 1

# Home vs Away goals comparison
if all(col in available_columns for col in ['FTHG', 'FTAG']):
    plt.subplot(3, 3, plot_count)
    goals_data = [df[available_columns['FTHG']], df[available_columns['FTAG']]]
    plt.boxplot(goals_data, labels=['Home Goals', 'Away Goals'])
    plt.title('Home vs Away Goals Distribution')
    plt.ylabel('Goals')
    plot_count += 1

# Shots analysis
if all(col in available_columns for col in ['HS', 'HST', 'AS', 'AST']):
    plt.subplot(3, 3, plot_count)
    plt.scatter(df[available_columns['HS']], df[available_columns['HST']], 
                alpha=0.6, label='Home', c='blue')
    plt.scatter(df[available_columns['AS']], df[available_columns['AST']], 
                alpha=0.6, label='Away', c='red')
    plt.xlabel('Total Shots')
    plt.ylabel('Shots on Target')
    plt.title('Shots vs Shots on Target')
    plt.legend()
    plot_count += 1

# Fouls analysis
if all(col in available_columns for col in ['HF', 'AF']):
    plt.subplot(3, 3, plot_count)
    fouls_data = [df[available_columns['HF']], df[available_columns['AF']]]
    plt.boxplot(fouls_data, labels=['Home Fouls', 'Away Fouls'])
    plt.title('Fouls Distribution')
    plt.ylabel('Number of Fouls')
    plot_count += 1

# Cards analysis
if all(col in available_columns for col in ['HY', 'AY', 'HR', 'AR']):
    plt.subplot(3, 3, plot_count)
    cards_data = [df[available_columns['HY']], df[available_columns['AY']], 
                  df[available_columns['HR']], df[available_columns['AR']]]
    plt.boxplot(cards_data, labels=['HY', 'AY', 'HR', 'AR'])
    plt.title('Cards Distribution')
    plt.ylabel('Number of Cards')
    plot_count += 1

# Corners analysis
if all(col in available_columns for col in ['HC', 'AC']):
    plt.subplot(3, 3, plot_count)
    corners_data = [df[available_columns['HC']], df[available_columns['AC']]]
    plt.boxplot(corners_data, labels=['Home Corners', 'Away Corners'])
    plt.title('Corners Distribution')
    plt.ylabel('Number of Corners')

plt.tight_layout()
plt.show()

# 2. Team Performance Analysis
print("\nüèÜ Analyzing Team Performance...")
if all(col in available_columns for col in ['HomeTeam', 'AwayTeam', 'FTR']):
    # Get all unique teams
    all_teams = list(set(df[available_columns['HomeTeam']].unique()) | 
                     set(df[available_columns['AwayTeam']].unique()))

    # Calculate team statistics
    team_stats = {}

    for team in all_teams:
        # Home games
        home_games = df[df[available_columns['HomeTeam']] == team]
        # Away games  
        away_games = df[df[available_columns['AwayTeam']] == team]
        
        # Total games
        total_games = len(home_games) + len(away_games)
        
        # Points calculation
        home_points = home_games[available_columns['FTR']].apply(
            lambda x: 3 if x == 'H' else (1 if x == 'D' else 0)).sum()
        away_points = away_games[available_columns['FTR']].apply(
            lambda x: 3 if x == 'A' else (1 if x == 'D' else 0)).sum()
        total_points = home_points + away_points
        
        # Goals (if available)
        goals_for = 0
        goals_against = 0
        if 'FTHG' in available_columns and 'FTAG' in available_columns:
            goals_for = home_games[available_columns['FTHG']].sum() + away_games[available_columns['FTAG']].sum()
            goals_against = home_games[available_columns['FTAG']].sum() + away_games[available_columns['FTHG']].sum()
        
        goal_difference = goals_for - goals_against
        
        # Wins, Draws, Losses
        wins = len(home_games[home_games[available_columns['FTR']] == 'H']) + \
               len(away_games[away_games[available_columns['FTR']] == 'A'])
        draws = len(home_games[home_games[available_columns['FTR']] == 'D']) + \
                len(away_games[away_games[available_columns['FTR']] == 'D'])
        losses = total_games - wins - draws
        
        team_stats[team] = {
            'Games': total_games,
            'Wins': wins,
            'Draws': draws,
            'Losses': losses,
            'Points': total_points,
            'GoalsFor': goals_for,
            'GoalsAgainst': goals_against,
            'GoalDifference': goal_difference,
            'WinRate': wins/total_games * 100 if total_games > 0 else 0
        }

    # Convert to DataFrame
    team_df = pd.DataFrame(team_stats).T
    team_df = team_df.sort_values('Points', ascending=False)

    print("\nüìã Premier League 2018-19 Standings:")
    print(team_df.head(10))

    # 3. Team Performance Visualizations
    print("\nüìà Creating Team Performance Charts...")
    plt.figure(figsize=(18, 12))

    # Top 10 teams by points
    plt.subplot(2, 3, 1)
    top_10 = team_df.head(10)
    plt.barh(top_10.index, top_10['Points'])
    plt.xlabel('Points')
    plt.title('Top 10 Teams by Points')
    plt.gca().invert_yaxis()

    # Goals scored vs conceded for top teams
    if 'GoalsFor' in team_df.columns and 'GoalsAgainst' in team_df.columns:
        plt.subplot(2, 3, 2)
        top_8 = team_df.head(8)
        x = np.arange(len(top_8))
        width = 0.35
        plt.bar(x - width/2, top_8['GoalsFor'], width, label='Goals For', alpha=0.7)
        plt.bar(x + width/2, top_8['GoalsAgainst'], width, label='Goals Against', alpha=0.7)
        plt.xlabel('Teams')
        plt.ylabel('Goals')
        plt.title('Goals For vs Against (Top 8)')
        plt.xticks(x, top_8.index, rotation=45)
        plt.legend()

    # Win rate distribution
    plt.subplot(2, 3, 3)
    plt.hist(team_df['WinRate'], bins=10, edgecolor='black')
    plt.xlabel('Win Rate (%)')
    plt.ylabel('Number of Teams')
    plt.title('Win Rate Distribution')

    # Home vs Away performance for top teams
    plt.subplot(2, 3, 4)
    top_6_teams = team_df.head(6).index
    home_away_data = []

    for team in top_6_teams:
        home_games = df[df[available_columns['HomeTeam']] == team]
        away_games = df[df[available_columns['AwayTeam']] == team]
        
        home_points = home_games[available_columns['FTR']].apply(
            lambda x: 3 if x == 'H' else (1 if x == 'D' else 0)).sum()
        away_points = away_games[available_columns['FTR']].apply(
            lambda x: 3 if x == 'A' else (1 if x == 'D' else 0)).sum()
        
        home_away_data.append([home_points, away_points])

    home_away_df = pd.DataFrame(home_away_data, index=top_6_teams, columns=['Home Points', 'Away Points'])
    home_away_df.plot(kind='bar', ax=plt.gca())
    plt.title('Home vs Away Points (Top 6 Teams)')
    plt.ylabel('Points')
    plt.xticks(rotation=45)

    # Goal difference visualization
    if 'GoalDifference' in team_df.columns:
        plt.subplot(2, 3, 5)
        colors = ['green' if x > 0 else 'red' for x in team_df['GoalDifference']]
        plt.barh(team_df.index, team_df['GoalDifference'], color=colors)
        plt.xlabel('Goal Difference')
        plt.title('Team Goal Differences')
        plt.gca().invert_yaxis()

    plt.tight_layout()
    plt.show()

# 4. Referee Analysis
print("\nüë®‚Äç‚öñÔ∏è Analyzing Referee Performance...")
if 'Referee' in available_columns:
    plt.figure(figsize=(15, 10))

    # Number of matches officiated by each referee
    plt.subplot(2, 2, 1)
    referee_counts = df[available_columns['Referee']].value_counts().head(15)
    plt.barh(referee_counts.index, referee_counts.values)
    plt.xlabel('Number of Matches')
    plt.title('Top 15 Referees by Matches Officiated')
    plt.gca().invert_yaxis()

    # Average cards per match by referee (if cards data available)
    if all(col in available_columns for col in ['HY', 'AY', 'HR', 'AR']):
        plt.subplot(2, 2, 2)
        referee_cards = df.groupby(available_columns['Referee']).agg({
            available_columns['HY']: 'mean', 
            available_columns['AY']: 'mean', 
            available_columns['HR']: 'mean', 
            available_columns['AR']: 'mean'
        }).mean(axis=1).sort_values(ascending=False).head(15)
        plt.barh(referee_cards.index, referee_cards.values)
        plt.xlabel('Average Cards per Match')
        plt.title('Top 15 Referees by Average Cards')
        plt.gca().invert_yaxis()

    # Home win percentage by referee
    plt.subplot(2, 2, 3)
    referee_home_win = df.groupby(available_columns['Referee']).apply(
        lambda x: (x[available_columns['FTR']] == 'H').sum() / len(x) * 100
    ).sort_values(ascending=False).head(15)
    plt.barh(referee_home_win.index, referee_home_win.values)
    plt.xlabel('Home Win Percentage (%)')
    plt.title('Referees with Highest Home Win %')
    plt.gca().invert_yaxis()

    # Average goals per match by referee
    if all(col in available_columns for col in ['FTHG', 'FTAG']):
        plt.subplot(2, 2, 4)
        referee_goals = df.groupby(available_columns['Referee']).apply(
            lambda x: (x[available_columns['FTHG']] + x[available_columns['FTAG']]).mean()
        ).sort_values(ascending=False).head(15)
        plt.barh(referee_goals.index, referee_goals.values)
        plt.xlabel('Average Goals per Match')
        plt.title('Referees with Highest Average Goals')
        plt.gca().invert_yaxis()

    plt.tight_layout()
    plt.show()

# 5. Monthly Performance and Trends
print("\nüìÖ Analyzing Monthly Trends...")
if 'Date' in df.columns:
    df['Month'] = df['Date'].dt.month
    df['Month_Name'] = df['Date'].dt.month_name()

    plt.figure(figsize=(15, 10))

    # Monthly goals trend
    if all(col in available_columns for col in ['FTHG', 'FTAG']):
        plt.subplot(2, 2, 1)
        monthly_goals = df.groupby('Month').agg({
            available_columns['FTHG']: 'mean', 
            available_columns['FTAG']: 'mean'
        })
        plt.plot(monthly_goals.index, monthly_goals[available_columns['FTHG']], 
                 marker='o', label='Home Goals')
        plt.plot(monthly_goals.index, monthly_goals[available_columns['FTAG']], 
                 marker='o', label='Away Goals')
        plt.xlabel('Month')
        plt.ylabel('Average Goals')
        plt.title('Monthly Average Goals Trend')
        plt.legend()
        plt.grid(True, alpha=0.3)

    # Monthly results distribution
    plt.subplot(2, 2, 2)
    monthly_results = pd.crosstab(df['Month'], df[available_columns['FTR']], normalize='index')
    monthly_results.plot(kind='bar', stacked=True, ax=plt.gca())
    plt.xlabel('Month')
    plt.ylabel('Proportion')
    plt.title('Monthly Result Distribution')
    plt.legend(title='Result')

    # Monthly cards trend
    if all(col in available_columns for col in ['HY', 'AY', 'HR', 'AR']):
        plt.subplot(2, 2, 3)
        monthly_cards = df.groupby('Month').agg({
            available_columns['HY']: 'mean', 
            available_columns['AY']: 'mean', 
            available_columns['HR']: 'mean', 
            available_columns['AR']: 'mean'
        }).sum(axis=1)
        plt.plot(monthly_cards.index, monthly_cards.values, marker='o', color='red')
        plt.xlabel('Month')
        plt.ylabel('Average Total Cards')
        plt.title('Monthly Cards Trend')
        plt.grid(True, alpha=0.3)

    # Monthly fouls trend
    if all(col in available_columns for col in ['HF', 'AF']):
        plt.subplot(2, 2, 4)
        monthly_fouls = df.groupby('Month').agg({
            available_columns['HF']: 'mean', 
            available_columns['AF']: 'mean'
        }).sum(axis=1)
        plt.plot(monthly_fouls.index, monthly_fouls.values, marker='o', color='green')
        plt.xlabel('Month')
        plt.ylabel('Average Total Fouls')
        plt.title('Monthly Fouls Trend')
        plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

# 6. Correlation Analysis
print("\nüîó Analyzing Statistical Correlations...")
# Select numerical columns for correlation
numerical_cols_to_use = [col for col in ['FTHG', 'FTAG', 'HTHG', 'HTAG', 'HS', 'AS', 'HST', 'AST', 
                  'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR'] 
                 if col in available_columns]

if len(numerical_cols_to_use) >= 2:
    plt.figure(figsize=(12, 10))
    correlation_matrix = df[[available_columns[col] for col in numerical_cols_to_use]].corr()
    mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
    sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', 
                center=0, square=True, fmt='.2f')
    plt.title('Correlation Matrix of Match Statistics')
    plt.tight_layout()
    plt.show()

    # Key insights from correlation
    print("\nüìä Key Correlations:")
    if all(col in available_columns for col in ['HS', 'FTHG']):
        print(f"Home Shots vs Home Goals: {df[available_columns['HS']].corr(df[available_columns['FTHG']]):.3f}")
    if all(col in available_columns for col in ['AS', 'FTAG']):
        print(f"Away Shots vs Away Goals: {df[available_columns['AS']].corr(df[available_columns['FTAG']]):.3f}")
    if all(col in available_columns for col in ['HST', 'FTHG']):
        print(f"Home Shots on Target vs Home Goals: {df[available_columns['HST']].corr(df[available_columns['FTHG']]):.3f}")
    if all(col in available_columns for col in ['AST', 'FTAG']):
        print(f"Away Shots on Target vs Away Goals: {df[available_columns['AST']].corr(df[available_columns['FTAG']]):.3f}")
    if all(col in available_columns for col in ['HF', 'HY']):
        print(f"Fouls vs Yellow Cards: {df[available_columns['HF']].corr(df[available_columns['HY']]):.3f}")

# 7. Final Summary and Key Insights
print("\n" + "=" * 60)
print("üèÜ PREMIER LEAGUE 2018-19 SEASON ANALYSIS SUMMARY")
print("=" * 60)

# Key statistics
total_matches = len(df)

if 'FTR' in available_columns:
    home_wins = len(df[df[available_columns['FTR']] == 'H'])
    away_wins = len(df[df[available_columns['FTR']] == 'A'])
    draws = len(df[df[available_columns['FTR']] == 'D'])
    home_win_percentage = (home_wins / total_matches) * 100

    print(f"üìä Season Overview:")
    print(f"   Total Matches: {total_matches}")
    print(f"   Home Wins: {home_wins} ({home_win_percentage:.1f}%)")
    print(f"   Away Wins: {away_wins} ({(away_wins/total_matches)*100:.1f}%)")
    print(f"   Draws: {draws} ({(draws/total_matches)*100:.1f}%)")

# Goals statistics
if all(col in available_columns for col in ['FTHG', 'FTAG']):
    total_goals = df[available_columns['FTHG']].sum() + df[available_columns['FTAG']].sum()
    avg_goals_per_match = total_goals / total_matches
    print(f"   Total Goals: {total_goals}")
    print(f"   Average Goals per Match: {avg_goals_per_match:.2f}")

# Top performers
if 'team_df' in locals():
    print(f"\nüèÖ Top 5 Teams:")
    top_5 = team_df.head()
    for i, (team, stats) in enumerate(top_5.iterrows(), 1):
        print(f"   {i}. {team}: {stats['Points']} pts, GD: {stats['GoalDifference']}")

# Most productive and leaky defenses
if 'team_df' in locals() and 'GoalsFor' in team_df.columns and 'GoalsAgainst' in team_df.columns:
    most_goals_for = team_df.nlargest(3, 'GoalsFor')
    most_goals_against = team_df.nlargest(3, 'GoalsAgainst')

    print(f"\n‚öΩ Best Attacks:")
    for team, stats in most_goals_for.iterrows():
        print(f"   {team}: {stats['GoalsFor']} goals")

    print(f"\nüö® Leakiest Defenses:")
    for team, stats in most_goals_against.iterrows():
        print(f"   {team}: {stats['GoalsAgainst']} goals conceded")

# Discipline statistics
if all(col in available_columns for col in ['HY', 'AY', 'HR', 'AR', 'HF', 'AF']):
    total_yellows = df[available_columns['HY']].sum() + df[available_columns['AY']].sum()
    total_reds = df[available_columns['HR']].sum() + df[available_columns['AR']].sum()
    total_fouls = df[available_columns['HF']].sum() + df[available_columns['AF']].sum()

    print(f"\nüü® Discipline Statistics:")
    print(f"   Total Yellow Cards: {total_yellows}")
    print(f"   Total Red Cards: {total_reds}")
    print(f"   Total Fouls: {total_fouls}")
    print(f"   Average Fouls per Match: {total_fouls/total_matches:.1f}")

# Referee statistics
if 'Referee' in available_columns:
    most_active_referee = df[available_columns['Referee']].value_counts().index[0]
    most_active_matches = df[available_columns['Referee']].value_counts().iloc[0]

    print(f"\nüë®‚Äç‚öñÔ∏è Referee Statistics:")
    print(f"   Most Active Referee: {most_active_referee} ({most_active_matches} matches)")

# Key insights
print(f"\nüí° KEY INSIGHTS:")
if 'FTR' in available_columns:
    print(f"   1. Home advantage: {home_win_percentage:.1f}% of matches won by home team")
if all(col in available_columns for col in ['HST', 'FTHG']):
    corr = df[available_columns['HST']].corr(df[available_columns['FTHG']])
    print(f"   2. Goal scoring efficiency: Shots on target correlate with goals (r={corr:.3f})")
if all(col in available_columns for col in ['HF', 'AF']):
    print(f"   3. Match intensity: Average of {total_fouls/total_matches:.1f} fouls per match")
if 'team_df' in locals():
    print(f"   4. Competitive balance: {len(team_df[team_df['WinRate'] > 50])} teams with >50% win rate")
if all(col in available_columns for col in ['FTHG', 'FTAG']):
    print(f"   5. Goal distribution: {avg_goals_per_match:.2f} goals per match on average")

print(f"\n‚úÖ Analysis complete! All available visualizations have been generated.")
print(f"üìÅ Data source: {github_url}")
```
