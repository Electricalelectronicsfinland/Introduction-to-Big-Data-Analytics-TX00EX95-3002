b) Explore the axis parameter:

```python
print("\n" + "="*60)
print("EXPLORING THE AXIS PARAMETER")
print("="*60)

# Default behavior (axis=0)
df_default = pd.concat([df_2018, df_2019])  # axis=0 is default
print(f"\n1. DEFAULT (axis=0) - Vertical concatenation:")
print(f"   Shape: {df_default.shape}")
print(f"   This stacks the dataframes on top of each other")
print(f"   Rows added: {df_2018.shape[0]} + {df_2019.shape[0]} = {df_default.shape[0]}")
print(f"   Columns remain: {df_default.shape[1]}")

# Explicit axis=0
df_axis0 = pd.concat([df_2018, df_2019], axis=0)
print(f"\n2. EXPLICIT axis=0 - Same as default:")
print(f"   Shape: {df_axis0.shape}")
print(f"   Result identical to default")

# axis=1 behavior
df_axis1 = pd.concat([df_2018, df_2019], axis=1)
print(f"\n3. axis=1 - Horizontal concatenation:")
print(f"   Shape: {df_axis1.shape}")
print(f"   This places dataframes side by side")
print(f"   Rows remain: {df_axis1.shape[0]} (uses the maximum row count)")
print(f"   Columns added: {df_2018.shape[1]} + {df_2019.shape[1]} = {df_axis1.shape[1]}")
print(f"   Note: Missing values (NaN) appear where row counts don't match")

print(f"\n4. COMPARISON:")
print(f"   Default parameter: axis=0")
print(f"   axis=0 â†’ Vertical stacking (adds rows)")
print(f"   axis=1 â†’ Horizontal concatenation (adds columns)")

# Show the actual data for axis=1
print(f"\n5. axis=1 result (first few columns):")
print(df_axis1.head())
```

Summary of answers:

The actual shape will depend on your specific dataframes

b) Axis parameter:
Default value: axis=0

axis=0: Vertical concatenation (stacks rows) - used for combining datasets with same columns

axis=1: Horizontal concatenation (side by side) - used for combining datasets with same rows but different columns

For football data across seasons, you typically use axis=0 (the default) to add more matches to your dataset

The key insight is that for combining multiple seasons of football data (where each season has the same columns but different matches), you want axis=0 (vertical concatenation), which is exactly what the default parameter gives you!

```python
import pandas as pd

print("=" * 80)
print("âš½ PREMIER LEAGUE 2018-19 GOAL ANALYSIS BY TEAM")
print("=" * 80)

# 1. Goals scored as Home Team
print("\n" + "ğŸ  " + "=" * 50)
print("ğŸ  HOME GOALS ANALYSIS")
print("ğŸ  " + "=" * 50)

groupedH = df1819.groupby('HomeTeam')['FTHG'].sum().sort_values(ascending=False)
print("ğŸ“ Total goals scored by each team as HOME team:")
print("-" * 55)
print(f"{'Team':<20} {'Home Goals':<12} {'Rank':<6}")
print("-" * 55)

for rank, (team, goals) in enumerate(groupedH.items(), 1):
    print(f"{team:<20} {goals:<12} #{rank:<5}")

# 2. Goals scored as Away Team
print("\n" + "âœˆï¸ " + "=" * 50)
print("âœˆï¸ AWAY GOALS ANALYSIS")
print("âœˆï¸ " + "=" * 50)

groupedA = df1819.groupby('AwayTeam')['FTAG'].sum().sort_values(ascending=False)
print("ğŸ“ Total goals scored by each team as AWAY team:")
print("-" * 55)
print(f"{'Team':<20} {'Away Goals':<12} {'Rank':<6}")
print("-" * 55)

for rank, (team, goals) in enumerate(groupedA.items(), 1):
    print(f"{team:<20} {goals:<12} #{rank:<5}")

# 3. Combined total goals
print("\n" + "ğŸ¯ " + "=" * 50)
print("ğŸ¯ TOTAL GOALS SCORED (HOME + AWAY)")
print("ğŸ¯ " + "=" * 50)

# Create a comprehensive summary
goal_summary = pd.DataFrame({
    'HomeGoals': df1819.groupby('HomeTeam')['FTHG'].sum(),
    'AwayGoals': df1819.groupby('AwayTeam')['FTAG'].sum()
}).fillna(0)

goal_summary['TotalGoals'] = goal_summary['HomeGoals'] + goal_summary['AwayGoals']
goal_summary_sorted = goal_summary.sort_values('TotalGoals', ascending=False)

print("ğŸ“ Complete goal breakdown for each team:")
print("-" * 70)
print(f"{'Team':<20} {'Home':<8} {'Away':<8} {'Total':<8} {'Rank':<6}")
print("-" * 70)

for rank, (team, row) in enumerate(goal_summary_sorted.iterrows(), 1):
    print(f"{team:<20} {row['HomeGoals']:<8} {row['AwayGoals']:<8} {row['TotalGoals']:<8} #{rank:<5}")

# 4. Multiple aggregations example
print("\n" + "ğŸ“Š " + "=" * 50)
print("ğŸ“Š ADVANCED GOAL STATISTICS")
print("ğŸ“Š " + "=" * 50)

print("ğŸ“ Home goals detailed analysis (sum, mean, max, min):")
home_detailed = df1819.groupby('HomeTeam')['FTHG'].agg(['sum', 'mean', 'max', 'min']).round(2)
home_detailed = home_detailed.sort_values('sum', ascending=False)
print(home_detailed)

print("\nğŸ“ Away goals detailed analysis (sum, mean, max, min):")
away_detailed = df1819.groupby('AwayTeam')['FTAG'].agg(['sum', 'mean', 'max', 'min']).round(2)
away_detailed = away_detailed.sort_values('sum', ascending=False)
print(away_detailed)

# 5. Top performers analysis
print("\n" + "ğŸ† " + "=" * 50)
print("ğŸ† TOP PERFORMERS ANALYSIS")
print("ğŸ† " + "=" * 50)

# Top 5 goal scorers
top_5_total = goal_summary_sorted.head(5)
print("ğŸ“ TOP 5 GOAL SCORING TEAMS:")
print("-" * 45)
for rank, (team, row) in enumerate(top_5_total.iterrows(), 1):
    print(f"#{rank} {team}: {row['TotalGoals']} goals "
          f"({row['HomeGoals']} home, {row['AwayGoals']} away)")

# Best home and away teams
best_home_team = groupedH.index[0]
best_away_team = groupedA.index[0]

print(f"\nğŸ“ BEST HOME ATTACK: {best_home_team} ({groupedH.iloc[0]} goals)")
print(f"ğŸ“ BEST AWAY ATTACK: {best_away_team} ({groupedA.iloc[0]} goals)")

# 6. Interesting insights
print("\n" + "ğŸ’¡ " + "=" * 50)
print("ğŸ’¡ INTERESTING INSIGHTS")
print("ğŸ’¡ " + "=" * 50)  # CORRECTED THIS LINE

# Home vs Away scoring efficiency
avg_home_goals = df1819['FTHG'].mean()
avg_away_goals = df1819['FTAG'].mean()

print(f"ğŸ“ League average goals per match:")
print(f"   â€¢ Home teams: {avg_home_goals:.2f} goals")
print(f"   â€¢ Away teams: {avg_away_goals:.2f} goals")
print(f"   â€¢ Home advantage: {avg_home_goals - avg_away_goals:.2f} goals")

# Most consistent scorers
home_consistency = df1819.groupby('HomeTeam')['FTHG'].std().sort_values()
away_consistency = df1819.groupby('AwayTeam')['FTAG'].std().sort_values()

print(f"\nğŸ“ Most consistent home scorers (lowest std dev):")
print(f"   1. {home_consistency.index[0]} (std: {home_consistency.iloc[0]:.2f})")
print(f"   2. {home_consistency.index[1]} (std: {home_consistency.iloc[1]:.2f})")

print(f"\nğŸ“ Most consistent away scorers (lowest std dev):")
print(f"   1. {away_consistency.index[0]} (std: {away_consistency.iloc[0]:.2f})")
print(f"   2. {away_consistency.index[1]} (std: {away_consistency.iloc[1]:.2f})")

# Additional insights
print(f"\nğŸ“ Goal scoring range:")
max_home_goals = df1819['FTHG'].max()
max_away_goals = df1819['FTAG'].max()
print(f"   â€¢ Highest home goals in a match: {max_home_goals}")
print(f"   â€¢ Highest away goals in a match: {max_away_goals}")
```
